```python
npy_dir = '/path/to/SpecVLA/observations.npy'
import numpy as np
observations_data = np.load(npy_dir, allow_pickle=True)
print(f"Observations data loaded from: {npy_dir}")
print(f"Observations data type: {type(observations_data)}")
print(f"Observations data keys: {list(observations_data.keys()) if isinstance(observations_data, dict) else 'N/A'}")
print(observations_data)
```

```python
# 从numpy数组中提取字典
obs_dict = observations_data.item()
print(f"数据类型: {type(obs_dict)}")
print(f"任务数量: {len(obs_dict)}")
print(f"任务ID列表: {list(obs_dict.keys())}")

# 查看第一个任务的结构
if len(obs_dict) > 0:
    first_task_id = list(obs_dict.keys())[0]
    print(f"\n任务 {first_task_id}:")
    print(f"  Episode数量: {len(obs_dict[first_task_id])}")
    
    # 查看第一个episode
    if len(obs_dict[first_task_id]) > 0:
        first_episode = list(obs_dict[first_task_id].keys())[0]
        print(f"  Episode {first_episode}:")
        print(f"    Observation数量: {len(obs_dict[first_task_id][first_episode])}")
        
        # 查看第一个observation的结构
        if len(obs_dict[first_task_id][first_episode]) > 0:
            first_obs = obs_dict[first_task_id][first_episode][0]
            print(f"    第一个observation的keys: {list(first_obs.keys())}")
            print(f"    full_image shape: {first_obs['full_image'].shape}")
            print(f"    state shape: {first_obs['state'].shape}")
```

```python
# 提取任务0的第一个轨迹
task_0_episode_0 = obs_dict[9][0]
print(f"任务0，Episode 0的轨迹长度: {len(task_0_episode_0)}")

state_list = [obs['state'] for obs in task_0_episode_0]
# 前三个维度是x,y,z
# print(state_list)
# 可视化前三个维度，学术一点，三维散点+连线图
```

```python
# 设置matplotlib backend
import matplotlib
matplotlib.use('Agg')  # 使用非交互式backend
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 提取xyz坐标（前三个维度）
states = np.array(state_list)
x = states[:, 0]
y = states[:, 1]
z = states[:, 2]

# 创建3D图
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# 绘制轨迹线
ax.plot(x, y, z, 'b-', linewidth=2, alpha=0.6, label='Trajectory')

# 绘制散点
scatter = ax.scatter(x, y, z, c=range(len(x)), cmap='viridis', 
                     s=50, alpha=0.8, edgecolors='black', linewidth=0.5)

# 标记起点和终点
ax.scatter(x[0], y[0], z[0], c='green', s=200, marker='o', 
          edgecolors='black', linewidth=2, label='Start', zorder=5)
ax.scatter(x[-1], y[-1], z[-1], c='red', s=200, marker='s', 
          edgecolors='black', linewidth=2, label='End', zorder=5)

# 设置标签
ax.set_xlabel('X Position (m)', fontsize=12, labelpad=10)
ax.set_ylabel('Y Position (m)', fontsize=12, labelpad=10)
ax.set_zlabel('Z Position (m)', fontsize=12, labelpad=10)

# 添加颜色条
cbar = plt.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)
cbar.set_label('Time Step', fontsize=11)

# 设置图例
ax.legend(loc='upper left', fontsize=11, framealpha=0.9)

# 设置网格
ax.grid(True, alpha=0.3)

# 设置视角
ax.view_init(elev=20, azim=45)

# 调整布局
plt.tight_layout()

# 保存图片而不是显示（避免backend问题）
save_path = '/path/to/SpecVLA/trajectory_plot.png'
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"图片已保存到: {save_path}")

print(f"\n轨迹统计:")
print(f"  X范围: [{x.min():.4f}, {x.max():.4f}]")
print(f"  Y范围: [{y.min():.4f}, {y.max():.4f}]")
print(f"  Z范围: [{z.min():.4f}, {z.max():.4f}]")
print(f"  轨迹长度: {len(x)} steps")

# 显示图片
from IPython.display import Image, display
display(Image(filename=save_path))
```
