```python
# ============================================
# Cell 3: 辅助函数 - 提取轨迹坐标和计算距离
# ============================================
def extract_trajectory_xyz(episode_data):
    """从episode数据中提取xyz轨迹坐标"""
    observations = episode_data['observations']
    states = np.array([obs['state'] for obs in observations])
    x = states[:, 0]
    y = states[:, 1]
    z = states[:, 2]
    return x, y, z

def compute_point_distances(traj1_xyz, traj2_xyz):
    """
    计算两条轨迹对应点之间的距离
    返回每个时间步的距离列表
    """
    x1, y1, z1 = traj1_xyz
    x2, y2, z2 = traj2_xyz
    
    # 取较短轨迹的长度
    min_len = min(len(x1), len(x2))
    
    distances = []
    for i in range(min_len):
        dist = np.sqrt((x1[i] - x2[i])**2 + (y1[i] - y2[i])**2 + (z1[i] - z2[i])**2)
        distances.append(dist)
    
    return np.array(distances)

def find_overlapping_points(distances, threshold=0.02):
    """
    找到距离小于阈值的重叠点索引
    threshold: 距离阈值（单位：米）
    """
    return np.where(distances < threshold)[0]

print("辅助函数定义完成")
```

```python
# ============================================
# Cell 3.5: 曲率计算函数
# ============================================
from scipy.interpolate import splprep, splev
from scipy.optimize import least_squares

def least_squares_circle_fit(points):
    """
    使用最小二乘法拟合圆，返回曲率（1/半径）
    points: (N, 3) 的点数组
    """
    if len(points) < 3:
        return np.nan
    
    # 使用3D点的投影到最佳拟合平面
    center = np.mean(points, axis=0)
    points_centered = points - center
    
    # SVD找到最佳平面
    _, _, vh = np.linalg.svd(points_centered)
    normal = vh[2, :]
    
    # 投影到平面
    points_2d = points_centered - np.outer(np.dot(points_centered, normal), normal)
    
    # 使用前两个主成分作为2D坐标
    u = vh[0, :]
    v = vh[1, :]
    x = np.dot(points_2d, u)
    y = np.dot(points_2d, v)
    
    # 拟合圆
    def calc_R(xc, yc):
        return np.sqrt((x - xc)**2 + (y - yc)**2)
    
    def f(c):
        Ri = calc_R(*c)
        return Ri - Ri.mean()
    
    center_estimate = np.array([x.mean(), y.mean()])
    result = least_squares(f, center_estimate)
    xc, yc = result.x
    Ri = calc_R(xc, yc)
    R = Ri.mean()
    
    # 曲率 = 1/半径
    curvature = 1.0 / R if R > 1e-6 else np.nan
    return curvature

def least_squares_circle_fit_radius(points):
    """
    使用最小二乘法拟合圆，直接返回曲率半径（半径R）
    points: (N, 3) 的点数组
    """
    if len(points) < 3:
        return np.nan
    
    # 使用3D点的投影到最佳拟合平面
    center = np.mean(points, axis=0)
    points_centered = points - center
    
    # SVD找到最佳平面
    _, _, vh = np.linalg.svd(points_centered)
    normal = vh[2, :]
    
    # 投影到平面
    points_2d = points_centered - np.outer(np.dot(points_centered, normal), normal)
    
    # 使用前两个主成分作为2D坐标
    u = vh[0, :]
    v = vh[1, :]
    x = np.dot(points_2d, u)
    y = np.dot(points_2d, v)
    
    # 拟合圆
    def calc_R(xc, yc):
        return np.sqrt((x - xc)**2 + (y - yc)**2)
    
    def f(c):
        Ri = calc_R(*c)
        return Ri - Ri.mean()
    
    center_estimate = np.array([x.mean(), y.mean()])
    result = least_squares(f, center_estimate)
    xc, yc = result.x
    Ri = calc_R(xc, yc)
    R = Ri.mean()
    
    # 直接返回半径R（曲率半径）
    return R if R > 1e-6 else np.nan

def compute_curvature_least_squares(x, y, z, window_size=5):
    """
    使用滑动窗口最小二乘圆拟合计算曲率
    """
    trajectory = np.column_stack([x, y, z])
    n = len(trajectory)
    curvatures = []
    
    for i in range(n):
        # 获取窗口
        start = max(0, i - window_size // 2)
        end = min(n, i + window_size // 2 + 1)
        
        if end - start < 3:
            curvatures.append(np.nan)
            continue
        
        window_points = trajectory[start:end]
        curvature = least_squares_circle_fit(window_points)
        curvatures.append(curvature)
    
    return np.array(curvatures)

def compute_radius_least_squares(x, y, z, window_size=5):
    """
    使用滑动窗口最小二乘圆拟合直接计算曲率半径
    """
    trajectory = np.column_stack([x, y, z])
    n = len(trajectory)
    radii = []
    
    for i in range(n):
        # 获取窗口
        start = max(0, i - window_size // 2)
        end = min(n, i + window_size // 2 + 1)
        
        if end - start < 3:
            radii.append(np.nan)
            continue
        
        window_points = trajectory[start:end]
        radius = least_squares_circle_fit_radius(window_points)
        radii.append(radius)
    
    return np.array(radii)

def compute_curvature_spline(x, y, z, smoothing=0):
    """
    使用样条插值后微分计算曲率
    """
    trajectory = np.column_stack([x, y, z])
    
    if len(trajectory) < 4:
        return np.full(len(trajectory), np.nan)
    
    # 参数化曲线长度
    distances = np.concatenate([[0], np.cumsum(np.linalg.norm(np.diff(trajectory, axis=0), axis=1))])
    
    try:
        # 样条插值 (k=3 为三次样条)
        tck, u = splprep([x, y, z], u=distances, s=smoothing, k=min(3, len(x)-1))
        
        # 计算一阶和二阶导数
        derivatives_1 = splev(u, tck, der=1)  # dx/du, dy/du, dz/du
        derivatives_2 = splev(u, tck, der=2)  # d²x/du², d²y/du², d²z/du²
        
        # 转换为numpy数组
        r_prime = np.array(derivatives_1).T  # (N, 3)
        r_double_prime = np.array(derivatives_2).T  # (N, 3)
        
        # 曲率公式: κ = |r' × r''| / |r'|³
        cross_product = np.cross(r_prime, r_double_prime)
        numerator = np.linalg.norm(cross_product, axis=1)
        denominator = np.linalg.norm(r_prime, axis=1) ** 3
        
        # 避免除零
        curvatures = np.where(denominator > 1e-10, numerator / denominator, np.nan)
        
        return curvatures
    except Exception as e:
        print(f"样条插值失败: {e}")
        return np.full(len(x), np.nan)

print("曲率计算函数定义完成")
```

```python
# ============================================
# Cell 3.6: 曲率可视化函数
# ============================================
def plot_curvature_comparison(retrieval_xyz, sd_xyz, 
                               method='least_squares',
                               window_size=5,
                               save_path=None):
    """
    绘制两条轨迹的曲率对比折线图
    
    参数:
    - retrieval_xyz: (x, y, z) 检索轨迹坐标
    - sd_xyz: (x, y, z) SD轨迹坐标
    - method: 'least_squares' 或 'spline'
    - window_size: 最小二乘法的窗口大小
    - save_path: 保存路径
    """
    r_x, r_y, r_z = retrieval_xyz
    s_x, s_y, s_z = sd_xyz
    
    # 计算曲率
    if method == 'least_squares':
        r_curvatures = compute_curvature_least_squares(r_x, r_y, r_z, window_size=window_size)
        s_curvatures = compute_curvature_least_squares(s_x, s_y, s_z, window_size=window_size)
        title_suffix = f'Least Squares (window={window_size})'
    elif method == 'spline':
        r_curvatures = compute_curvature_spline(r_x, r_y, r_z)
        s_curvatures = compute_curvature_spline(s_x, s_y, s_z)
        title_suffix = 'Spline Interpolation'
    else:
        raise ValueError(f"Unknown method: {method}")
    
    # 创建图
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # 绘制检索轨迹的曲率
    steps_r = np.arange(len(r_curvatures))
    valid_r = ~np.isnan(r_curvatures)
    ax.plot(steps_r[valid_r], r_curvatures[valid_r], 'b-', linewidth=2, 
            label='Retrieval', alpha=0.8)
    
    # 绘制SD轨迹的曲率
    steps_s = np.arange(len(s_curvatures))
    valid_s = ~np.isnan(s_curvatures)
    ax.plot(steps_s[valid_s], s_curvatures[valid_s], color='darkorange', 
            linewidth=2, label='SD', alpha=0.8)
    
    # 图表设置
    ax.set_xlabel('Time Step', fontsize=14, fontweight='bold')
    ax.set_ylabel('Curvature (1/m)', fontsize=14, fontweight='bold')
    ax.set_title(f'Trajectory Curvature Comparison - {title_suffix}', 
                 fontsize=16, fontweight='bold', pad=15)
    ax.legend(loc='upper right', fontsize=12, framealpha=0.9, edgecolor='black')
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # 统计信息
    r_mean = np.nanmean(r_curvatures)
    r_std = np.nanstd(r_curvatures)
    s_mean = np.nanmean(s_curvatures)
    s_std = np.nanstd(s_curvatures)
    
    # 添加统计文本
    stats_text = f'Retrieval: μ={r_mean:.4f}, σ={r_std:.4f}\nSD: μ={s_mean:.4f}, σ={s_std:.4f}'
    ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, 
            fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    
    # 保存
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"曲率图已保存: {save_path}")
    
    plt.close()
    return save_path

print("曲率可视化函数定义完成")
```

```python
# ============================================
# Cell 3.7: 曲率半径可视化函数
# ============================================
def plot_curvature_radius_comparison(retrieval_xyz, sd_xyz, 
                                      method='least_squares',
                                      window_size=5,
                                      save_path=None):
    """
    绘制两条轨迹的曲率半径对比折线图（归一化）
    
    参数:
    - retrieval_xyz: (x, y, z) 检索轨迹坐标
    - sd_xyz: (x, y, z) SD轨迹坐标
    - method: 'least_squares' 或 'spline'
    - window_size: 最小二乘法的窗口大小
    - save_path: 保存路径
    """
    r_x, r_y, r_z = retrieval_xyz
    s_x, s_y, s_z = sd_xyz
    
    # 计算曲率半径
    if method == 'least_squares':
        # 直接计算半径，不需要先算曲率再取倒数
        r_radius = compute_radius_least_squares(r_x, r_y, r_z, window_size=window_size)
        s_radius = compute_radius_least_squares(s_x, s_y, s_z, window_size=window_size)
        title_suffix = f'Least Squares (window={window_size})'
    elif method == 'spline':
        # 样条插值先计算曲率，再取倒数得到半径
        r_curvatures = compute_curvature_spline(r_x, r_y, r_z)
        s_curvatures = compute_curvature_spline(s_x, s_y, s_z)
        r_radius = np.where((r_curvatures > 1e-6) & (~np.isnan(r_curvatures)), 
                            1.0 / r_curvatures, np.nan)
        s_radius = np.where((s_curvatures > 1e-6) & (~np.isnan(s_curvatures)), 
                            1.0 / s_curvatures, np.nan)
        title_suffix = 'Spline Interpolation'
    else:
        raise ValueError(f"Unknown method: {method}")
    
    # 归一化曲率半径
    r_radius_norm = minmax_normalize(r_radius)
    s_radius_norm = minmax_normalize(s_radius)
    
    # 创建图
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # 从第5个点开始绘制（跳过前5个点）
    start_idx = 5
    
    # 绘制检索轨迹的归一化曲率半径
    steps_r = np.arange(len(r_radius_norm))
    valid_r = ~np.isnan(r_radius_norm)
    # 同时满足：索引>=start_idx 且 valid
    valid_r = valid_r & (steps_r >= start_idx)
    ax.plot(steps_r[valid_r], r_radius_norm[valid_r], 'b-', linewidth=4, alpha=0.8, marker='^', markersize=7)
    
    # 绘制SD轨迹的归一化曲率半径 - 同一张图用同一种marker 'o'
    steps_s = np.arange(len(s_radius_norm))
    valid_s = ~np.isnan(s_radius_norm)
    # 同时满足：索引>=start_idx 且 valid
    valid_s = valid_s & (steps_s >= start_idx)
    ax.plot(steps_s[valid_s], s_radius_norm[valid_s], color='darkorange', linewidth=4, alpha=0.8, marker='^', markersize=7)
    
    # 移除轴标签、标题、图例、刻度标签，但保留grid和spines
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_title('')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.legend().set_visible(False)
    ax.grid(True, alpha=0.3, linestyle='--')
    
    plt.tight_layout()
    
    # 保存
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"曲率半径图已保存（归一化）: {save_path}")
    
    plt.close()
    return save_path

print("曲率半径可视化函数定义完成")
```

```python
# ============================================
# Cell 3.8: 位移指标和加权平均计算函数
# ============================================
def compute_displacement_metric(x, y, z, window_size=5):
    """
    计算位移指标：在滑动窗口内，窗口最后一个点与前面所有点的欧式距离之和
    
    参数:
    - x, y, z: 轨迹坐标
    - window_size: 滑动窗口大小
    
    返回:
    - displacement_metrics: 每个时间步的位移指标
    """
    trajectory = np.column_stack([x, y, z])
    n = len(trajectory)
    displacement_metrics = []
    
    for i in range(n):
        # 获取窗口：从 max(0, i-window_size+1) 到 i（包含i）
        # 这样窗口的最后一个点就是当前点i
        start = max(0, i - window_size + 1)
        end = i + 1  # 包含当前点
        
        if end - start < 2:
            displacement_metrics.append(np.nan)
            continue
        
        window_points = trajectory[start:end]
        # 窗口最后一个点（当前点）
        last_point = window_points[-1]
        
        # 计算最后一个点与前面所有点的欧式距离之和
        total_distance = 0.0
        for j in range(len(window_points) - 1):  # 不包括最后一个点自己
            dist = np.linalg.norm(last_point - window_points[j])
            total_distance += dist
        
        displacement_metrics.append(total_distance)
    
    return np.array(displacement_metrics)

def minmax_normalize(values):
    """
    MinMax归一化
    
    参数:
    - values: 需要归一化的数组
    
    返回:
    - normalized_values: 归一化后的数组
    """
    valid_values = values[~np.isnan(values)]
    if len(valid_values) == 0:
        return values
    
    min_val = np.nanmin(values)
    max_val = np.nanmax(values)
    
    if max_val - min_val < 1e-10:
        # 如果所有值都相同，返回全0或全1
        return np.where(~np.isnan(values), 0.5, np.nan)
    
    normalized = (values - min_val) / (max_val - min_val)
    return normalized

def compute_weighted_metric(radius, displacement, weight_r=0.5):
    """
    计算加权平均指标
    
    参数:
    - radius: 曲率半径数组
    - displacement: 位移指标数组
    - weight_r: 曲率半径的权重（默认0.5）
    
    返回:
    - weighted_metric: 加权后的指标
    """
    # 归一化
    radius_norm = minmax_normalize(radius)
    displacement_norm = minmax_normalize(displacement)
    
    # 加权平均
    weighted = weight_r * radius_norm + (1 - weight_r) * displacement_norm
    
    return weighted, radius_norm, displacement_norm

print("位移指标和加权平均计算函数定义完成")
```

```python
# ============================================
# Cell 3.7: 曲率半径可视化函数
# ============================================
def plot_curvature_radius_comparison(retrieval_xyz, sd_xyz, 
                                      method='least_squares',
                                      window_size=5,
                                      save_path=None):
    """
    绘制两条轨迹的曲率半径对比折线图（归一化）
    
    参数:
    - retrieval_xyz: (x, y, z) 检索轨迹坐标
    - sd_xyz: (x, y, z) SD轨迹坐标
    - method: 'least_squares' 或 'spline'
    - window_size: 最小二乘法的窗口大小
    - save_path: 保存路径
    """
    r_x, r_y, r_z = retrieval_xyz
    s_x, s_y, s_z = sd_xyz
    
    # 计算曲率半径
    if method == 'least_squares':
        # 直接计算半径，不需要先算曲率再取倒数
        r_radius = compute_radius_least_squares(r_x, r_y, r_z, window_size=window_size)
        s_radius = compute_radius_least_squares(s_x, s_y, s_z, window_size=window_size)
        title_suffix = f'Least Squares (window={window_size})'
    elif method == 'spline':
        # 样条插值先计算曲率，再取倒数得到半径
        r_curvatures = compute_curvature_spline(r_x, r_y, r_z)
        s_curvatures = compute_curvature_spline(s_x, s_y, s_z)
        r_radius = np.where((r_curvatures > 1e-6) & (~np.isnan(r_curvatures)), 
                            1.0 / r_curvatures, np.nan)
        s_radius = np.where((s_curvatures > 1e-6) & (~np.isnan(s_curvatures)), 
                            1.0 / s_curvatures, np.nan)
        title_suffix = 'Spline Interpolation'
    else:
        raise ValueError(f"Unknown method: {method}")
    
    # 归一化曲率半径
    r_radius_norm = minmax_normalize(r_radius)
    s_radius_norm = minmax_normalize(s_radius)
    
    # 创建图
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # 从第5个点开始绘制（跳过前5个点）
    start_idx = 5
    
    # 绘制检索轨迹的归一化曲率半径
    steps_r = np.arange(len(r_radius_norm))
    valid_r = ~np.isnan(r_radius_norm)
    # 同时满足：索引>=start_idx 且 valid
    valid_r = valid_r & (steps_r >= start_idx)
    ax.plot(steps_r[valid_r], r_radius_norm[valid_r], 
            'b-', linewidth=4, alpha=0.8, 
            marker='^', markersize=7, 
            markerfacecolor='blue', markeredgecolor='darkblue', markeredgewidth=1)
    
    # 绘制SD轨迹的归一化曲率半径
    steps_s = np.arange(len(s_radius_norm))
    valid_s = ~np.isnan(s_radius_norm)
    # 同时满足：索引>=start_idx 且 valid
    valid_s = valid_s & (steps_s >= start_idx)
    ax.plot(steps_s[valid_s], s_radius_norm[valid_s], 
            color='darkorange', linewidth=4, alpha=0.8, 
            marker='^', markersize=7, 
            markerfacecolor='darkorange', markeredgecolor='orangered', markeredgewidth=1)
    
    # 移除轴标签、标题、图例、刻度标签，但保留grid和spines
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_title('')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.legend().set_visible(False)
    ax.grid(True, alpha=0.3, linestyle='--')
    
    plt.tight_layout()
    
    # 保存
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"曲率半径图已保存（归一化）: {save_path}")
    
    plt.close()
    return save_path

print("曲率半径可视化函数定义完成")
```

```python
# ============================================
# Cell 4: 找到符合条件的轨迹对（交叉配对版本）
# ============================================

# 先打印所有可用的任务名称
print("数据中可用的任务：")
print("="*60)
for task_id in retrieval_info.keys():
    for eid in retrieval_info[task_id].keys():
        task_desc = retrieval_info[task_id][eid]['task_description']
        print(f"  Task {task_id}: {task_desc}")
        break  # 每个task只打印一次
print("="*60)

# ========== 任务筛选设置 ==========
# 设置为 None 则处理所有任务，设置为具体关键词则只处理包含该关键词的任务
# 注意：使用原始任务描述中的关键词，如 "bowl on top of the cabinet"（不是下划线格式）
TARGET_TASK_NAME = "push the plate to the front of the stove"  # 使用原始描述中的关键词
# TARGET_TASK_NAME = None  # 取消注释这行并注释上一行可处理所有任务
# ==================================

def find_trajectory_pairs(retrieval_info, sd_info, target_task_name=None):
    """
    找到不同成功状态组合的轨迹对（同一任务内交叉配对）
    
    参数：
    - target_task_name: 可选，指定只处理包含该字符串的任务（不区分大小写）
    
    返回：
    - both_success: 两种方法都成功的 (task_id, retrieval_episode_idx, sd_episode_idx, task_desc) 列表
    - sd_success_retrieval_fail: SD成功但检索失败的列表
    - retrieval_success_sd_fail: 检索成功但SD失败的列表
    - both_fail: 两种方法都失败的列表
    """
    both_success = []
    sd_success_retrieval_fail = []
    retrieval_success_sd_fail = []
    both_fail = []
    
    for task_id in retrieval_info.keys():
        if task_id not in sd_info:
            continue
        
        # 收集该任务下不同成功状态的episodes
        retrieval_success_episodes = []
        retrieval_fail_episodes = []
        sd_success_episodes = []
        sd_fail_episodes = []
        
        task_desc = None
        
        # 分类检索结果
        for r_eid in retrieval_info[task_id].keys():
            if task_desc is None:
                task_desc = retrieval_info[task_id][r_eid]['task_description']
            
        # 筛选任务：如果指定了target_task_name，则只处理包含该字符串的任务
        if target_task_name is not None and task_desc is not None:
            if target_task_name.lower() not in task_desc.lower():
                continue
        
        # 重新分类（因为上面的循环只是为了获取task_desc）
        for r_eid in retrieval_info[task_id].keys():
            if retrieval_info[task_id][r_eid]['success']:
                retrieval_success_episodes.append(r_eid)
            else:
                retrieval_fail_episodes.append(r_eid)
        
        # 分类SD结果
        for s_eid in sd_info[task_id].keys():
            if sd_info[task_id][s_eid]['success']:
                sd_success_episodes.append(s_eid)
            else:
                sd_fail_episodes.append(s_eid)
        
        # 交叉配对
        # 1. 两种方法都成功：检索成功 × SD成功
        for r_eid in retrieval_success_episodes:
            for s_eid in sd_success_episodes:
                both_success.append((task_id, r_eid, s_eid, task_desc))
        
        # 2. SD成功但检索失败：检索失败 × SD成功
        for r_eid in retrieval_fail_episodes:
            for s_eid in sd_success_episodes:
                sd_success_retrieval_fail.append((task_id, r_eid, s_eid, task_desc))
        
        # 3. 检索成功但SD失败：检索成功 × SD失败
        for r_eid in retrieval_success_episodes:
            for s_eid in sd_fail_episodes:
                retrieval_success_sd_fail.append((task_id, r_eid, s_eid, task_desc))
        
        # 4. 两种方法都失败：检索失败 × SD失败
        for r_eid in retrieval_fail_episodes:
            for s_eid in sd_fail_episodes:
                both_fail.append((task_id, r_eid, s_eid, task_desc))
    
    return both_success, sd_success_retrieval_fail, retrieval_success_sd_fail, both_fail

# 找到轨迹对（交叉配对）
both_success, sd_success_retrieval_fail, retrieval_success_sd_fail, both_fail = find_trajectory_pairs(retrieval_info, sd_info, target_task_name=TARGET_TASK_NAME)

if TARGET_TASK_NAME:
    print(f"仅筛选任务: {TARGET_TASK_NAME}")

print(f"统计结果（交叉配对）:")
print(f"  两种方法都成功: {len(both_success)} 对")
print(f"  SD成功但检索失败: {len(sd_success_retrieval_fail)} 对")
print(f"  检索成功但SD失败: {len(retrieval_success_sd_fail)} 对")
print(f"  两种方法都失败: {len(both_fail)} 对")

# 显示部分示例
if len(both_success) > 0:
    print(f"\n两种方法都成功的示例 (前3个):")
    for i, (tid, r_eid, s_eid, desc) in enumerate(both_success[:3]):
        print(f"  Task {tid}, Retrieval Episode {r_eid} vs SD Episode {s_eid}: {desc[:50]}...")

if len(sd_success_retrieval_fail) > 0:
    print(f"\nSD成功但检索失败的示例 (前3个):")
    for i, (tid, r_eid, s_eid, desc) in enumerate(sd_success_retrieval_fail[:3]):
        print(f"  Task {tid}, Retrieval Episode {r_eid} vs SD Episode {s_eid}: {desc[:50]}...")
```

```python
# ============================================
# Cell 8: 批量绘制多个对比图（含曲率和曲率半径分析）
# ============================================
def batch_plot_comparisons(trajectory_pairs, retrieval_data, sd_data, 
                           category_name, category_short_name, max_plots=5):
    """批量绘制多个轨迹对比图（含曲率和曲率半径分析）"""
    print(f"\n{'='*60}")
    print(f"批量绘制: {category_name}")
    print(f"{'='*60}")
    
    if len(trajectory_pairs) == 0:
        print(f"没有找到 {category_name} 的轨迹对")
        return []
    
    saved_paths = []
    num_to_plot = min(len(trajectory_pairs), max_plots)
    
    for i, (task_id, r_eid, s_eid, task_desc) in enumerate(trajectory_pairs[:num_to_plot]):
        print(f"\n[{i+1}/{num_to_plot}] Task {task_id}, Retrieval Episode {r_eid} vs SD Episode {s_eid}")
        
        # 提取轨迹
        r_episode = retrieval_data[task_id][r_eid]
        s_episode = sd_data[task_id][s_eid]
        
        r_xyz = extract_trajectory_xyz(r_episode)
        s_xyz = extract_trajectory_xyz(s_episode)
        
        r_success = retrieval_info[task_id][r_eid]['success']
        s_success = sd_info[task_id][s_eid]['success']
        
        # 生成文件名和路径
        task_name = get_task_name_from_description(task_desc)
        
        # 按任务名创建文件夹，然后在其下创建情况子文件夹
        task_folder = os.path.join(output_dir, task_name)
        category_folder = os.path.join(task_folder, category_short_name)
        os.makedirs(category_folder, exist_ok=True)
        
        # 文件名格式：任务名_情况_r检索episode_s模型episode.png
        filename = f"{task_name}_{category_short_name}_r{r_eid}_s{s_eid}.png"
        save_path = os.path.join(category_folder, filename)
        
        # 绘制3D轨迹图
        plot_trajectory_comparison(
            r_xyz, s_xyz,
            retrieval_success=r_success, sd_success=s_success,
            task_description=task_desc,
            task_id=task_id, episode_idx=f"R{r_eid}_S{s_eid}",
            overlap_threshold=0.02,
            save_path=save_path
        )
        saved_paths.append(save_path)
        
        # ========== 绘制曲率半径分析图 ==========
        # 最小二乘法曲率半径
        radius_ls_path = save_path.replace('.png', '_radius_squares.png')
        plot_curvature_radius_comparison(
            r_xyz, s_xyz,
            method='least_squares',
            window_size=5,
            save_path=radius_ls_path
        )
        saved_paths.append(radius_ls_path)
        
        # ========== 绘制位移指标分析图 ==========
        displacement_path = save_path.replace('.png', '_displacement.png')
        plot_displacement_metric_comparison(
            r_xyz, s_xyz,
            window_size=5,
            save_path=displacement_path
        )
        saved_paths.append(displacement_path)
        
        # ========== 绘制加权指标分析图 ==========
        # 加权指标 (r=0.5)
        weighted_path = save_path.replace('.png', '_weighted_r0.5.png')
        plot_weighted_metric_comparison(
            r_xyz, s_xyz,
            window_size=5,
            weight_r=0.5,
            save_path=weighted_path
        )
        saved_paths.append(weighted_path)
    
    return saved_paths

# 批量绘制各类别的图
print("\n开始批量绘制...")

# 1. 两种方法都成功 
paths_both_success = batch_plot_comparisons(
    both_success, retrieval_data, sd_data, 
    "SD Success & Retrieval Success", "both_success", max_plots=100
)

# 2. SD成功但检索失败
paths_sd_success = batch_plot_comparisons(
    sd_success_retrieval_fail, retrieval_data, sd_data, 
    "SD Success & Retrieval Fail", "sd_success_retrieval_fail", max_plots=100
)

# 3. 检索成功但SD失败
paths_retrieval_success = batch_plot_comparisons(
    retrieval_success_sd_fail, retrieval_data, sd_data, 
    "SD Fail & Retrieval Success", "sd_fail_retrieval_success", max_plots=100
)

# 4. 两种方法都失败
paths_both_fail = batch_plot_comparisons(
    both_fail, retrieval_data, sd_data, 
    "SD Fail & Retrieval Fail", "both_fail", max_plots=100
)

print(f"\n\n{'='*60}")
print("批量绘制完成!")
print(f"图片保存目录: {output_dir}")
print(f"{'='*60}")
total_pairs = len(both_success) + len(sd_success_retrieval_fail) + len(retrieval_success_sd_fail) + len(both_fail)
print(f"\n总计生成图片:")
print(f"  - 3D轨迹图: {total_pairs} 张")
print(f"  - 曲率半径图: {total_pairs} 张")
print(f"  - 位移指标图: {total_pairs} 张")
print(f"  - 加权指标图: {total_pairs} 张")
print(f"  - 总计: {4 * total_pairs} 张")
```

```python
# ============================================
# Cell 9: 总体统计和汇总
# ============================================
def compute_overall_statistics(trajectory_pairs, retrieval_data, sd_data, category_name):
    """计算某类轨迹对的总体统计"""
    if len(trajectory_pairs) == 0:
        return None
    
    all_avg_distances = []
    all_overlap_ratios = []
    
    for task_id, r_eid, s_eid, _ in trajectory_pairs:
        r_episode = retrieval_data[task_id][r_eid]
        s_episode = sd_data[task_id][s_eid]
        
        r_xyz = extract_trajectory_xyz(r_episode)
        s_xyz = extract_trajectory_xyz(s_episode)
        
        distances = compute_point_distances(r_xyz, s_xyz)
        overlap_indices = find_overlapping_points(distances, threshold=0.02)
        
        all_avg_distances.append(np.mean(distances))
        overlap_ratio = len(overlap_indices) / len(distances) if len(distances) > 0 else 0
        all_overlap_ratios.append(overlap_ratio)
    
    return {
        'category': category_name,
        'count': len(trajectory_pairs),
        'avg_distance_mean': np.mean(all_avg_distances),
        'avg_distance_std': np.std(all_avg_distances),
        'overlap_ratio_mean': np.mean(all_overlap_ratios) * 100,
        'overlap_ratio_std': np.std(all_overlap_ratios) * 100
    }

# 计算统计
print("="*70)
print("总体统计汇总")
print("="*70)

stats_list = []

stats = compute_overall_statistics(both_success, retrieval_data, sd_data, "SD Success & Retrieval Success")
if stats:
    stats_list.append(stats)

stats = compute_overall_statistics(sd_success_retrieval_fail, retrieval_data, sd_data, "SD Success & Retrieval Fail")
if stats:
    stats_list.append(stats)

stats = compute_overall_statistics(retrieval_success_sd_fail, retrieval_data, sd_data, "SD Fail & Retrieval Success")
if stats:
    stats_list.append(stats)

stats = compute_overall_statistics(both_fail, retrieval_data, sd_data, "SD Fail & Retrieval Fail")
if stats:
    stats_list.append(stats)

# 打印统计表
print(f"\n{'Category':<40} {'Count':>8} {'Avg Dist (m)':>15} {'Overlap %':>12}")
print("-"*70)
for s in stats_list:
    print(f"{s['category']:<40} {s['count']:>8} {s['avg_distance_mean']:>10.4f}±{s['avg_distance_std']:.4f} {s['overlap_ratio_mean']:>8.1f}±{s['overlap_ratio_std']:.1f}%")

print("\n" + "="*70)
print("分析结论:")
print("="*70)
print("- 两种方法都成功时，轨迹重叠度高说明两种方法产生了相似的路径")
print("- SD成功但检索失败时，可以观察SD如何纠正错误的检索结果")
print("- 检索成功但SD失败时，说明检索效果好但模型生成能力不足")
print("- 两种方法都失败时，可能是任务本身难度较大")

# 保存统计结果到文件
analysis_file = os.path.join(output_dir, "Analysis.txt")
with open(analysis_file, 'w', encoding='utf-8') as f:
    f.write("="*70 + "\n")
    f.write("总体统计汇总\n")
    f.write("="*70 + "\n\n")
    
    f.write(f"{'Category':<40} {'Count':>8} {'Avg Dist (m)':>15} {'Overlap %':>12}\n")
    f.write("-"*70 + "\n")
    for s in stats_list:
        f.write(f"{s['category']:<40} {s['count']:>8} {s['avg_distance_mean']:>10.4f}±{s['avg_distance_std']:.4f} {s['overlap_ratio_mean']:>8.1f}±{s['overlap_ratio_std']:.1f}%\n")
    
    f.write("\n" + "="*70 + "\n")
    f.write("分析结论:\n")
    f.write("="*70 + "\n")
    f.write("- 两种方法都成功时，轨迹重叠度高说明两种方法产生了相似的路径\n")
    f.write("- SD成功但检索失败时，可以观察SD如何纠正错误的检索结果\n")
    f.write("- 检索成功但SD失败时，说明检索效果好但模型生成能力不足\n")
    f.write("- 两种方法都失败时，可能是任务本身难度较大\n")
    
    f.write("\n" + "="*70 + "\n")
    f.write("详细统计:\n")
    f.write("="*70 + "\n")
    f.write(f"SD Success & Retrieval Success: {len(both_success)} 对\n")
    f.write(f"SD Success & Retrieval Fail: {len(sd_success_retrieval_fail)} 对\n")
    f.write(f"SD Fail & Retrieval Success: {len(retrieval_success_sd_fail)} 对\n")
    f.write(f"SD Fail & Retrieval Fail: {len(both_fail)} 对\n")
    f.write(f"总计: {len(both_success) + len(sd_success_retrieval_fail) + len(retrieval_success_sd_fail) + len(both_fail)} 对\n")

print(f"\n统计结果已保存到: {analysis_file}")
```

```python
# ============================================
# Cell 10.8: 计算和绘制混合指标分布
# ============================================
print("="*60)
print("计算混合指标 (位移指标 1:1 曲率半径)")
print("="*60)

# 定义归一化范围
DISP_MIN = 0.000009
DISP_MAX = 0.123381
RADIUS_MIN = 0.000001
RADIUS_MAX = 0.014989

print(f"位移指标归一化范围: [{DISP_MIN}, {DISP_MAX}]")
print(f"曲率半径归一化范围: [{RADIUS_MIN}, {RADIUS_MAX}]")

# 收集所有混合指标
all_hybrid_values = []

print(f"\n总计 {len(all_pairs)} 对轨迹")

from tqdm import tqdm

for task_id, r_eid, s_eid, task_desc in tqdm(all_pairs, desc="计算混合指标"):
    # ========== 检索轨迹 ==========
    r_episode = retrieval_data[task_id][r_eid]
    r_xyz = extract_trajectory_xyz(r_episode)
    
    # 计算原始指标
    r_displacement = compute_displacement_metric(r_xyz[0], r_xyz[1], r_xyz[2], window_size=window_size)
    r_radius = compute_radius_least_squares(r_xyz[0], r_xyz[1], r_xyz[2], window_size=window_size)
    
    # Min-Max 归一化
    r_disp_norm = np.clip((r_displacement - DISP_MIN) / (DISP_MAX - DISP_MIN), 0, 1)
    r_radius_norm = np.clip((r_radius - RADIUS_MIN) / (RADIUS_MAX - RADIUS_MIN), 0, 1)
    
    # 计算混合指标 (1:1)
    r_hybrid = 0.5 * r_disp_norm + 0.5 * r_radius_norm
    
    # 收集有效值
    all_hybrid_values.extend(r_hybrid[~np.isnan(r_hybrid)])
    
    # ========== SD轨迹 ==========
    s_episode = sd_data[task_id][s_eid]
    s_xyz = extract_trajectory_xyz(s_episode)
    
    # 计算原始指标
    s_displacement = compute_displacement_metric(s_xyz[0], s_xyz[1], s_xyz[2], window_size=window_size)
    s_radius = compute_radius_least_squares(s_xyz[0], s_xyz[1], s_xyz[2], window_size=window_size)
    
    # Min-Max 归一化
    s_disp_norm = np.clip((s_displacement - DISP_MIN) / (DISP_MAX - DISP_MIN), 0, 1)
    s_radius_norm = np.clip((s_radius - RADIUS_MIN) / (RADIUS_MAX - RADIUS_MIN), 0, 1)
    
    # 计算混合指标 (1:1)
    s_hybrid = 0.5 * s_disp_norm + 0.5 * s_radius_norm
    
    # 收集有效值
    all_hybrid_values.extend(s_hybrid[~np.isnan(s_hybrid)])

# 转换为numpy数组
all_hybrid_values = np.array(all_hybrid_values)

print(f"\n数据收集完成!")
print(f"混合指标样本数: {len(all_hybrid_values)}")

# ========== 统计信息 ==========
print("\n" + "="*60)
print("混合指标统计")
print("="*60)
print(f"最小值: {np.min(all_hybrid_values):.6f}")
print(f"最大值: {np.max(all_hybrid_values):.6f}")
print(f"均值: {np.mean(all_hybrid_values):.6f}")
print(f"中位数 (50%): {np.median(all_hybrid_values):.6f}")
print(f"标准差: {np.std(all_hybrid_values):.6f}")
print(f"25%分位数: {np.percentile(all_hybrid_values, 25):.6f}")
print(f"75%分位数: {np.percentile(all_hybrid_values, 75):.6f}")
print(f"90%分位数: {np.percentile(all_hybrid_values, 90):.6f}")
print(f"95%分位数: {np.percentile(all_hybrid_values, 95):.6f}")
print(f"99%分位数: {np.percentile(all_hybrid_values, 99):.6f}")

# ========== 设置保存路径 ==========
save_dir = "/path/to/SpecVLA/vis_distri"
os.makedirs(save_dir, exist_ok=True)

# ========== 绘制混合指标分布 ==========
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# 1. 箱线图
ax1 = axes[0]
bp = ax1.boxplot([all_hybrid_values], vert=True, patch_artist=True,
                   labels=['Hybrid Metric (1:1)'],
                   boxprops=dict(facecolor='lightgreen', alpha=0.7),
                   medianprops=dict(color='red', linewidth=2),
                   whiskerprops=dict(linewidth=1.5),
                   capprops=dict(linewidth=1.5))

# 设置横轴和纵轴标签
ax1.set_xlabel('Metric Type', fontsize=16, fontweight='bold')
ax1.set_ylabel('Hybrid Metric Value (Normalized)', fontsize=16, fontweight='bold')
ax1.set_title('Hybrid Metric Distribution - Box Plot\n(Displacement 1:1 Curvature Radius)', 
              fontsize=18, fontweight='bold', pad=15)
ax1.grid(True, alpha=0.3, axis='y')
ax1.tick_params(axis='both', labelsize=14)

# 添加统计文本（备注）
hybrid_stats = f'Min: {np.min(all_hybrid_values):.4f}\n25%: {np.percentile(all_hybrid_values, 25):.4f}\nMedian: {np.median(all_hybrid_values):.4f}\n75%: {np.percentile(all_hybrid_values, 75):.4f}\n90%: {np.percentile(all_hybrid_values, 90):.4f}\nMax: {np.max(all_hybrid_values):.4f}'
ax1.text(0.02, 0.98, hybrid_stats, transform=ax1.transAxes, 
         fontsize=12, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# 添加图注（备注说明）
# ax1.text(0.5, -0.12, 
#          'Note: Box shows IQR (25%-75%), red line is median, whiskers extend to 1.5×IQR',
#          transform=ax1.transAxes, fontsize=12, ha='center', style='italic',
#          bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

# 2. 直方图
ax2 = axes[1]
n, bins, patches = ax2.hist(all_hybrid_values, bins=50, color='lightgreen', 
                             alpha=0.7, edgecolor='black')

# 添加分位数线
percentiles = [25, 50, 75, 90]
colors = ['blue', 'green', 'orange', 'red']
for pct, color in zip(percentiles, colors):
    pct_val = np.percentile(all_hybrid_values, pct)
    ax2.axvline(pct_val, color=color, linestyle='--', linewidth=2.5, 
                label=f'{pct}th Percentile: {pct_val:.4f}')

ax2.axvline(np.mean(all_hybrid_values), color='purple', linestyle='-', 
            linewidth=2.5, label=f'Mean: {np.mean(all_hybrid_values):.4f}')

# 设置横轴和纵轴标签
ax2.set_xlabel('Hybrid Metric Value (Normalized)', fontsize=16, fontweight='bold')
ax2.set_ylabel('Frequency (Count)', fontsize=16, fontweight='bold')
ax2.set_title('Hybrid Metric Distribution - Histogram\n(Displacement 1:1 Curvature Radius)', 
              fontsize=18, fontweight='bold', pad=15)
ax2.legend(fontsize=12, loc='upper right', framealpha=0.9)
ax2.grid(True, alpha=0.3, axis='y')
ax2.tick_params(axis='both', labelsize=14)

# 添加图注（备注说明）
# ax2.text(0.5, -0.12, 
#          f'Note: Total samples: {len(all_hybrid_values):,}, Bins: 50, Dashed lines show percentiles',
#          transform=ax2.transAxes, fontsize=12, ha='center', style='italic',
#          bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)  # 为底部备注留出空间

# 保存到指定路径
hybrid_path = os.path.join(save_dir, "Hybrid_Metric_Distribution.png")
plt.savefig(hybrid_path, dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
print(f"\n混合指标分布图已保存: {hybrid_path}")
plt.close()

# 显示图表
display(Image(filename=hybrid_path))

# ========== 保存混合指标统计结果 ==========
hybrid_stats_file = os.path.join(save_dir, "Hybrid_Metric_Statistics.txt")
with open(hybrid_stats_file, 'w', encoding='utf-8') as f:
    f.write("="*60 + "\n")
    f.write("混合指标统计 (位移指标 1:1 曲率半径)\n")
    f.write("="*60 + "\n\n")
    
    f.write("归一化范围:\n")
    f.write("-"*60 + "\n")
    f.write(f"位移指标: [{DISP_MIN}, {DISP_MAX}]\n")
    f.write(f"曲率半径: [{RADIUS_MIN}, {RADIUS_MAX}]\n\n")
    
    f.write("统计信息:\n")
    f.write("-"*60 + "\n")
    f.write(f"样本数: {len(all_hybrid_values)}\n")
    f.write(f"最小值: {np.min(all_hybrid_values):.6f}\n")
    f.write(f"最大值: {np.max(all_hybrid_values):.6f}\n")
    f.write(f"均值: {np.mean(all_hybrid_values):.6f}\n")
    f.write(f"标准差: {np.std(all_hybrid_values):.6f}\n\n")
    
    f.write("分位数:\n")
    f.write("-"*60 + "\n")
    f.write(f"25%分位数: {np.percentile(all_hybrid_values, 25):.6f}\n")
    f.write(f"50%分位数 (中位数): {np.percentile(all_hybrid_values, 50):.6f}\n")
    f.write(f"75%分位数: {np.percentile(all_hybrid_values, 75):.6f}\n")
    f.write(f"90%分位数: {np.percentile(all_hybrid_values, 90):.6f}\n")
    f.write(f"95%分位数: {np.percentile(all_hybrid_values, 95):.6f}\n")
    f.write(f"99%分位数: {np.percentile(all_hybrid_values, 99):.6f}\n")

print(f"混合指标统计结果已保存到: {hybrid_stats_file}")
```
